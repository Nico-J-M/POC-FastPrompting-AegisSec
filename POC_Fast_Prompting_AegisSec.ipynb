## 3) Fast Prompting ‚Äî Framework pr√°ctico (CRISPE/RAFT)

**CRISPE / RAFT (adaptado):**
- **R**ole (Rol): define la pericia del modelo.
- **A**ction (Acci√≥n): lo que debe producir exactamente.
- **F**ormat (Formato): listas, JSON, headings, etc.
- **T**one (Tono) + **C**onstraints (Restricciones): largo, estilo, palabras prohibidas/obligatorias, p√∫blico objetivo.
- **E**xamples (Ejemplos): bocetos, contra‚Äëejemplos, muestras.
- **P**arameters (Par√°metros): temperatura, longitud, etc. (opcional, si usas API).

**Plus:** *Generate ‚Üí Critique ‚Üí Refine* (auto‚Äëcr√≠tica con r√∫brica).
